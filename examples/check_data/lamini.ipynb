```python
import json
```

```python
d = open('lamini_dataset.txt', 'r')
```

```python
dlist = d.readlines()
```

```python
len(dlist)
```

```python
qlist = [json.loads(t) for t in dlist]
```

```python
len(qlist)
```

```python
for q in qlist:
    qk = q.keys()
    for k in qk:
        if k != 'response' and k != 'question':
            print(k)
```

```python
for i in range(100):
    print('\n' + str(i))
    print('Question\n' + qlist[i]['question'])
    print('Response\n' + qlist[i]['response'])
```
